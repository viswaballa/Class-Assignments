```python
# -*- coding: utf-8 -*-
"""
Student Activity: Market Capitalization.

This script showcases the use of Python Dicts to determine the
bank names associated with the corresponding market cap ranges.
"""

# Banks and Market Caps
#-----------------------
# JP Morgan Chase: 327
# Bank of America: 302
# Citigroup: 173
# Wells Fargo: 273
# Goldman Sachs: 87
# Morgan Stanley: 72
# U.S. Bancorp: 83
# TD Bank: 108
# PNC Financial Services: 67
# Capital One: 47
# FNB Corporation: 4
# First Hawaiian Bank: 3
# Ally Financial: 12
# Wachovia: 145
# Republic Bancorp: .97

# @TODO: Initialize a dictionary of banks and market caps (in billions)

banks = {"JP Morgan Chase": 327, "Bank of America": 302, "Citigroup": 173, "Wells Fargo": 273,
"Goldman Sachs": 87, "Morgan Stanley": 72, "U.S. Bancorp": 83, "TD Bank": 108,
"PNC Financial Services": 67, "Capital One": 47, "FNB Corporation": 4, "First Hawaiian Bank": 3,
"Ally Financial": 12, "Wachovia": 145, "Republic Bancorp": .97}


# @TODO: Change the market cap for 'Citigroup'

banks['Citigroup'] = 90

# @TODO: Add a new bank and market cap pair

banks['American Express'] = 75

# @TODO: Remove a bank from the dictionary

del banks['Wachovia']


# @TODO: Print the modified dictionary

print(banks)


#Initialize metric variables

total_market_cap = 0
bank_count = 0
average = 0

# Initialize minimum and maximum key value pairs

minimum_key = ""
minimum_value = 0

maximum_key = ""
maximum_value = 0 

# Market cap lists 
mega_cap = []
large_cap = []
mid_cap = []
small_cap = []

# Key value pairs of the dictionary 
print()
for bank_name, market_cap in banks.items():
    print(f"Name: {bank_name} | Market Cap: {market_cap}")
    
    # Calculate sum of market cap 
    total_market_cap += int(market_cap)
    bank_count += 1
    
    # Determine min value associated by key
    if minimum_value == 0:
        minimum_value = market_cap
    elif market_cap < minimum_value: 
        minimum_value = market_cap
        minimum_key = bank_name
    
    # Determine max value associated by key
    if market_cap > maximum_value: 
        maximum_value = market_cap
        maximum_key = bank_name
    
    # Group banks by market cap categories 
    if market_cap >= 300: 
        mega_cap.append(bank_name)
    elif market_cap >= 10 and market_cap < 300: 
        large_cap.append(bank_name)
    elif market_cap >= 2 and market_cap < 10: 
        mid_cap.append(bank_name)
    elif market_cap < 2: 
        small_cap.append(bank_name)
    
print()


# Calculate avg market cap of banks
avg_market_cap = round(total_market_cap/bank_count,2)

# Print the metrics: 

print(f"Total market capitalization: {total_market_cap}")
print(f"Total number of banks: {bank_count}")
print(f"Average market capitalization: {avg_market_cap}")
print(f"Largest Bank: {maximum_key}")
print(f"Smallest Bank: {minimum_key}")
print("---------------------------------------------------")
print(f"Mega cap banks: {mega_cap}")
print(f"Large cap banks: {large_cap}")
print(f"Mid cap banks: {mid_cap}")
print(f"Small cap banks: {small_cap}")
print() 

```
